```python
import pandas as pd
import re
import os
df = pd.read_csv("data/raw_feedback.csv")
df.set_index("CustomerID", inplace=True)
df
```

```python
df.dropna(subset=["Feedback"], inplace=True)
df.drop_duplicates(subset=["Feedback"], inplace=True)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"[^\w\s]", "", text) 
    text = re.sub(r"\s+", " ", text).strip() 
    return text
```

```python
df["CleanedFeedback"] = df["Feedback"].apply(clean_text)
df.to_csv("data/cleaned_feedback.csv")
print("Cleaning Done!")
print(f"\nTotal Feedbacks After Cleaning: {len(df)}")
print(f"Duplicates Removed: {25 - len(df)}")
print("\nSample Before vs After Cleaning:\n")
sample = df[["Feedback", "CleanedFeedback"]].head(5)
print(sample.to_string(index=False))
```
